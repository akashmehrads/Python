```python
#12. Classify the employees into 3 categories.
#    Category 1 : employees below 18.5 bmi - underweight
#    Category 2 : employees with bmi in the range of 18.5 to 24.9 - fit
#    Category 3 : employee with bmi over 24.9 - overweight

# What is the average expenditure of employees in each category.
# Display the bar chart for the same
```

```python
df12=dfemp
```

```python
#Classify the employees into 3 categories.
for i in range (0,len(df12)):
    
    if(df12.iloc[i]['bmi']<18.5):
        df12.Category[i]='underweight'
    
    elif((df12.iloc[i]['bmi']<=18.5) and (df12.iloc[i]['bmi']<24.9)):
        df12.Category[i]='fit'
    
    else:
        df12.Category[i]='overweight'
```

```python
df12.head()
```

```python
#What is the average expenditure of employees in each category.
print("Average expenditure in each category: ")
dfcat_avgexp=pd.pivot_table(df12,index='Category',values='expenditure',aggfunc='mean')
dfcat_avgexp
```

```python
# Bar chart for average expenditure of employees in each category
plt.bar(x=dfcat_avgexp.index, height=dfcat_avgexp.expenditure,color=['red','green','blue'])
plt.title("Average Expenditure of employees in each category")
plt.xlabel("Category")
plt.ylabel("Expenditure")
plt.show()
```
