4.For the Titanic Dataset, Create Pivot table for the following functionalities:
    
    
    1. Use Sex and Pclass as index parameter for pivot table and apply median function on Survived
    2. Use Sex and Embarked as index parameter for pivot table and apply mean function on Fare   
    3. Use Sex and Embarked as index parameter for pivot table and apply min,max,mean function on Fare



```python
df1=pd.read_csv("train_titanic.csv")
df1.head()
```

```python
df1_1=pd.pivot_table(df1,index=['Sex','Pclass'],values='Survived',aggfunc='median')
df1_1
```

```python
df1_2=pd.pivot_table(df1,index=['Sex','Embarked'],values='Fare',aggfunc='mean')
df1_2
```

```python
df1_3=pd.pivot_table(df1,index=['Sex','Embarked'],values='Fare',aggfunc=['min','max','mean'])
df1_3
```

```python
#mean of surv, max of fare
df1_4=pd.pivot_table(df1,index=['Sex','Embarked'],values=['Survived','Fare'],aggfunc={'Survived':'mean','Fare':'max'})
df1_4
```
