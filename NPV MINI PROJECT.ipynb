```python
#d-What is the average expenditure of employees in each category.

#FOR CATEGORY UNDER WEIGHT:
mean=df_under_weight.expenditure.mean()
print('Average expenditure of Under Weight employees= ',mean)
print()
#FOR CATEGORY FIT:
mean1=df_fit.expenditure.mean()
print('Average expenditure of Fit employees= ',mean1)
print()
#FOR CATEGORY OVER WEIGHT:
mean2=df_over_weight.expenditure.mean()
print('Average expenditure of Over Weight employees= ',mean2)
```
